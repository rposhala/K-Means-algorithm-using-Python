# Implementation of K- Means algorithm using python to cluster the raw data

```
def Kmeans(df,K):
  i = np.random.choice(len(df),K,replace=False)
  centriods = df[i]
  dis = np.zeros((len(df),K))
  prev_wcss = 0
  for j in range(200):
    clusters = []
    for i in range(len(centriods)):
      dis[:,i] = np.linalg.norm(centriods[i]-df,axis = 1)
    labels = np.argmin(dis,axis=1)
    for i in range(len(centriods)):
      cluster = np.asarray([df[np.where(labels == i)[0]]])
      clusters.extend(cluster)
      centriods[i] = np.mean(df[np.where(labels == i)[0]],axis=0)
    wcss = WCSS(clusters)
    if -0.01 <= prev_wcss - wcss <= 0.01:
      # print(j,prev_wcss - wcss)
      break
    prev_wcss = wcss
    # prev_clusters = clusters
  return clusters
   
```

# Function to calculate Within cluster sum of squares

```
def WCSS(y):
  wcss = 0
  for i in y:
    centre = np.mean(i,axis = 0)
    wcss += np.sum((np.linalg.norm(i-centre,axis=1))*2)
  # print(wcss)
  return wcss
```
